```python
from syndalib import synth
```

```python
mycircle = synth.circle(radius=2,center=(2,2),n=240)
```

```python
mycircle.shape
import matplotlib.pyplot as plt

plt.gca().set_aspect('equal', adjustable='box')
plt.scatter(mycircle[:,0], mycircle[:,1], s=5)
```

```python
e = synth.ellipse(3,6,(2,2),100)
plt.gca().set_aspect('equal', adjustable='box')
plt.scatter(e[:,0], e[:,1], s=5)
```

```python
p = synth.parabola(1, 2, 5, 400, outliers_perc=0.5)
plt.gca().set_aspect('equal', adjustable='box')
plt.scatter(p[:,0], p[:,1], s=5)
```

```python
h = synth.hyperbola(5,7,(4,5),400, outliers_perc= 0.4)
plt.gca().set_aspect('equal', adjustable='box')
plt.scatter(h[:,0], h[:,1], s=5)
```

```python
l = synth.line((3,4), (8,5), 400, outliers_perc=0.5)
plt.gca().set_aspect('equal', adjustable='box')
plt.scatter(l[:,0], l[:,1], s=5)
```
